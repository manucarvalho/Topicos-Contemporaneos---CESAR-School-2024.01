```python
import random
from langchain.tools import tool

@tool
def add(a: int, b: int) -> int:
    """Soma dois inteiros.

    Args:
        a: Primeiro inteiro
        b: Segundo inteiro
    """
    return a + b

@tool
def multiply(a: int, b: int) -> int:
    """Multiplica dois inteiros.

    Args:
        a: Primeiro inteiro
        b: Segundo inteiro
    """
    return a * b

@tool
def get_weather(city: str) -> str:
    """Retorna um clima aleatório para uma cidade.

    Args:
        city: Nome da cidade
    """
    weather = random.choice(["ensolarado", "nublado", "chuvoso", "tempestuoso", "neve"])
    return weather

tools = [add, multiply, get_weather]
llm_with_tools = llm.bind_tools(tools)
```

```python
get_weather.invoke({"city": "Recife-RN"})
```

```python
from langchain_core.messages import HumanMessage

messages = [HumanMessage("Qual a previsão do tempo em Recife-RN? E também quanto é 123*321?")]

ai_msg = llm_with_tools.invoke(messages)

print(f"ContentString: {ai_msg.content}")
print(f"ToolCalls: {ai_msg.tool_calls}")
```

### Exercício 1
Adicione uma função nas ferramentas do agente para criar arquivos de texto. Em seguida, teste o agente.

```python
@tool
def create_txt(name, text):
    """Cria um arquivo de texto

    Args:
        name: Nome do arquivo
        text: Conteúdo do arquivo
    """
    with open(name, 'w') as f:
        f.write(text)

tools = [add, multiply, get_weather, create_txt]
```

```python
agent_executor = create_react_agent(llm, tools)
response = agent_executor.invoke({
    "messages": [
        HumanMessage("Qual a previsão do tempo em Natal-RN? E também crie um arquivo de texto com o nome teste.txt sobre a importância da etapa de teste no desenvolvimento de um sistema.")
    ]
})

print(response["messages"][-1].content)
```

### Exercício 2
Crie um agente que conte a quantidade de ocorrências de uma determinada letra em uma palavra. Em seguida, teste o agente.

```python
@tool
def count_letters(text, letter):
    """Conta o número de ocorrências de uma letra em um texto.

    Args:
        text (str): O texto a ser analisado.
        letter (str): A letra a ser contada.

    Returns:
        int: O número de ocorrências da letra no texto.
    """
    return text.count(letter)

tools_2 = [count_letters]
```

```python
from langgraph.prebuilt import create_react_agent

agent_executor_2 = create_react_agent(llm, tools_2)

response = agent_executor_2.invoke({
    "messages": [
        HumanMessage("Quantas vezes a letra a aparece na palavra Melancia?")
    ]
})

print(response["messages"][-1].content)
```
